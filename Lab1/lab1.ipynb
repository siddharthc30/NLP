```python
class File:
    
    def count_lines(self, file):
        self.file = file
        a = open(file, "r")
        a = a.read()
        lines = len(a.split("\n"))
        print(lines)
        
    
    def clean_file(self, file):
        self.file = file
        t = open(file, "r")
        t = t.read()
        
        text = [re.sub(r"[^a-zA-Z]+", ' ', t) for t in t.split()]
        text = " ".join(text)
        text = text.lower()
        return text
        #print(text)
        
    def count_words(self, text):
        count_words = len(text.split())
        #print(text.split())
        print(count_words)
        
    def count_charac(self, text):
        words = text.split()
        count = 0
        for w in words:
            count = count + len(w)
        
        print(count)
        return words
        
        
    def vocab(self, words):
        vocabulory = {}
        for w in words:
            if w in vocabulory:
                vocabulory[w] += 1
            else:
                vocabulory.update({w: 1})
        for x in vocabulory:
            print(x)
        
        return vocabulory
    
    def freq(self, vocabulory):
        for k, v in vocabulory.items():
            print(k, ":", v)
        
```

```python
query = input("Enter the query ")
query = query.split()
def similarity(vocab, query):
    count = 0
    for w in query:
        if w in vocab:
            count = vocab[w]
    return count

total_d = {
    'f1':similarity(vocab1, query),
    'f2':similarity(vocab2, query),
    'f3':similarity(vocab3, query),
    'f4':similarity(vocab4, query),
    'f5':similarity(vocab5, query),
}

decreasing = sorted(total_d.items(),key=lambda x:x[1],reverse = True)
for i in decreasing:
    print("The file ",i[0]," had a similarity :",i[1])
    
```
